```python
# CHOOSING COLOR MAP FOR COLOR BARS
# HERE WE WANT TO SHOW DIFFERENCE BETWEEN DIFFERENT COLORS SCALES
from matplotlib.colors import LinearSegmentedColormap
def grey_scale(cmap):
    '''Return grey scale of any given color map'''
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    
    # Convert RGBA to perceived grayscale luminace
    RGB_weight = [0.299, 0.587, 0.114]
    luminace = np.sqrt(np.dot(colors[:, :3] ** 2, RGB_weight))
    colors[:, :3] = luminace[:, np.newaxis]
    
    return LinearSegmentedColormap.from_list(cmap.name + '_grey', colors, cmap.N)
def view_colormap(cmap):
    '''Plot a colormap with its grayscale equivalent'''
    cmap = plt.cm.get_cmap(cmap)
    colors = cmap(np.arange(cmap.N))
    cmap = grey_scale(cmap)
    grayscale = cmap(np.arange(cmap.N))
    
    fix, ax = plt.subplots(2, figsize=(6, 2),
                           subplot_kw=dict(xticks=[], yticks=[]))
    ax[0].imshow([colors], extent=[0, 10, 0, 1])
    ax[1].imshow([grayscale], extent=[0, 10, 0, 1])
# TEST THOSE FUNCTION ON CMAP EXAMPLES
view_colormap('jet')
view_colormap('viridis')
view_colormap('cubehelix')
view_colormap('RdBu')
view_colormap('Blues')
```

```python
# HISTOGRAMS OF SUBSETS
tips = sns.load_dataset('tips')
tips['tips_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row='sex', col='time', margin_titles=True)
grid.map(plt.hist, 'tips_pct', bins=np.linspace(0, 40, 15))
```

```python
# FACTOR PLOTS
with sns.axes_style(style='ticks'):
    g = sns.factorplot('day', 'total_bill', 'sex', data=tips, kind='box')
    g.set_axis_labels('Day', 'Total Bill')
```

```python
# JOIN DISTRBUTIONS - HEX scatter + hist
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip', data=tips, kind='hex')
```

```python
# SAME BUT WITH STYLE='REG' IT GIVES ADDITIONAL INSIGHT
with sns.axes_style('white'):
    sns.jointplot('total_bill', 'tip', data=tips, kind='reg')
```
