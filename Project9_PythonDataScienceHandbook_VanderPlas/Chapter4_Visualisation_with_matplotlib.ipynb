```python
# example 
plt.style.use('classic')
x = np.linspace(0, 10, 100)
fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
plt.show
```

```python
# TWO PLOTS ON ONE FIGURE
# method1: matlab
plt.figure()        # create a plot figure
plt.subplot(2, 1, 1)    # (rows, columns, panel number)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
# method2: Python BETTER
fig, ax = plt.subplots(2)      # first create grid of plots

ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
# multiple lines and different colors
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x - 0), color='blue')
ax.plot(x, np.sin(x - 1), color='g')
ax.plot(x, np.sin(x - 2), color='0.75')
ax.plot(x, np.sin(x - 3), color='#FFDD44')
ax.plot(x, np.sin(x - 4), color=(1.0, 0.2, 0.3))
ax.plot(x, np.sin(x - 5), color='chartreuse')
```

```python
# multiple lines and different line styles
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
ax.plot(x, np.sin(x - 0), linestyle='-')
ax.plot(x, np.sin(x - 1), linestyle='--')
ax.plot(x, np.sin(x - 2), linestyle='-.')
ax.plot(x, np.sin(x - 3), linestyle=':')
```

```python
# MUTIPLE LINES WITH DIFFERENT COLORS AND STYLES
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1, 10, 1000)
ax.plot(x, np.sin(x), '-k')
ax.plot(x, np.sin(x + 1), '-.c')
ax.plot(x, np.sin(x + 2), '--c')
ax.plot(x, np.sin(x + 3), ':c')
```

```python
# Adjusting X and y limits
fig = plt.figure()
ax = plt.axes()

x = np.linspace(1, 10, 1000)

plt.axis([0, 11, -1.5, 1.5])
ax.plot(x, np.sin(x))
```

```python
# SAVING LABELS, TITLES, LEGENDS
#fig = plt.figure()
ax = plt.axes()

ax.plot(x, np.sin(x))
plt.title('A Sinus curve', size=20)
plt.xlabel('x')
plt.ylabel('sin(x)')
```

```python
# ADD LEGEND TO PLOT
# style used
plt.style.use('classic')
# set up
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0, 10, 1000)
# fill grid
ax.plot(x, np.sin(x), '-k', label='sin(x)')
ax.plot(x, np.sin(x + 1), '-.c', label='sin(x+1)')
ax.plot(x, np.sin(x + 2), '--c', label='sin(x+2)')
ax.plot(x, np.sin(x + 3), ':c', label='sin(x+3)')
# Appearance
ax.set(xlim=(0, 11),
       xlabel='x', ylabel='sin(x)',
       title='Sinus plot')
ax.axis('equal')
# Show Legend
ax.legend(loc='upper center', ncol=2, fancybox=True, framealpha=1, shadow=True, borderpad=1)
```

```python
# Building scatter plot
fig = plt.figure()
ax = plt.axes()

x = np.linspace(0, 10, 100)
y = np.sin(x)

ax.plot(x, y, 'oc', label='sin(x)')
```

```python
# PLOT WITH LINES AND SCATTER POINTS + Adidtional specification
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1, 10, 30)
y = np.sin(x)
ax.plot(x, y, '-d', label='sin(x)', 
        color='k', linewidth=2,
        markerfacecolor='k', markersize=10, 
        markeredgecolor='c', markeredgewidth=3)
ax.set(xlim=(1, 11), ylim=(-1.5, 1.5),
       title='Sin(x)')
ax.legend()
```

```python
# ERROR BARS
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)
ax.plot(x, y)

ax.errorbar(x, y, yerr=dy, fmt='.k')
```

```python
# ERROR BARS 2 
fig = plt.figure()
ax = plt.axes()
x = np.linspace(1, 10, 100)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(100)
#ax.plot(x, y)

ax.errorbar(x, y, yerr=dy, fmt='o', color='k',
            ecolor='lightgrey', elinewidth=3, capsize=0)
```

```python
# CONTOUR PLOTS
x = np.linspace(0, 25, 500)
y = np.linspace(0, 25, 500)

def f(x, y):
    return np.sin(x) + np.cos(y)

X, Y = np.meshgrid(x, y)

Z = f(X, Y)

plt.contourf(X, Y, Z, 100, cmap='RdGy')
plt.colorbar()
#fix = plt.figure()
#ax = plt.axes()
```

```python
# BETTER WAY TO DO TWO DIMENSION FILLED SCATTER PLOTS
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower', cmap='RdGy')
plt.colorbar()
plt.axis(aspect='image')
plt.show()
```
