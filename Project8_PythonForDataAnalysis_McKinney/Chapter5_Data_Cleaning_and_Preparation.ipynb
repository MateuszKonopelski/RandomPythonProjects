### Task: Computing Dummies on multiple categories in column

Method for building dummies column based on numerous of criteria. Effective for small data sets. 


```python
# set up env
import numpy as np
import pandas as pd
```

```python
# 1) CREATE EXAMPLE DF
df = pd.DataFrame({'A': np.random.rand(10),
                   'B': np.random.choice(['a', 'b', 'c', 'd'], 10),
                   'C': ['C1|C2|C3|C4', 'C2', 'C3', 'C4', 'C1',
                         'C1|C2', 'C2|C3', 'C1|C3|C4', 'C4|C3', 'C1|C3']})
df
```

```python
# 2) Create a list of all possible categories in Cc
categories = []
for rows in df.C:
    categories.extend(rows.split('|'))
categories = pd.unique(categories)
categories
```

```python
# 3) create empty zeros matrix with the lenght of original table and columns of categories 
zero_matrix = np.zeros((len(df), len(categories)))
dummies = pd.DataFrame(zero_matrix, columns=categories, dtype=np.int)
dummies
```

```python
# 4) loop through df to fill the dummies table
for i, cat in enumerate(df.C):
    indices = dummies.columns.get_indexer(cat.split('|'))
    dummies.iloc[i, indices] = 1
dummies
```

```python
# 5) join with original table
df.join(dummies.add_prefix('Cat_')).drop('C', axis=1)
```
