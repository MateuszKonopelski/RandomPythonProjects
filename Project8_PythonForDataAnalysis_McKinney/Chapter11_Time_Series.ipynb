```python
# et up env
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
% matplotlib inline
```

```python
# Create df
dir = 'C:\\Users\\Lenovo\\PycharmProjects\\pandas_exercises\\09_Time_Series\\Apple_Stock\\'
data = pd.read_csv(dir + 'appl_1980_2014.csv', parse_dates=True)
data.head()
```

```python
# FORMAT DATE TO DATETIME AND SET IT AS INDEX
data.Date = pd.to_datetime(data.Date)
data.index = data.Date
data.drop('Date', axis=1, inplace=True)
data.sort_index(inplace=True)
```

```python
# GENERATING DATE RANGES
pd.date_range(start='2017-01-01', end='2017-01-07')
pd.date_range(start='2017-01-01', periods=5)
pd.date_range(end='2017-01-01', periods=5)
pd.date_range(start='2017-01-01', end='2017-12-01', freq='BM')
pd.date_range('2018-04-15', '2018-04-22', freq='B')
pd.date_range('2018-04-15 15:03:20', periods=5)
pd.date_range('2018-04-15 15:03:20', periods=5, normalize=True)
pd.date_range('2018-04-19 15:20', '2018-04-19 22:59', freq='2h')
pd.date_range('2018-04-19 15:20', '2018-04-19 22:59', freq='1h30min')
pd.date_range('2018-01-01', periods=12, freq='WOM-3FRI')    # third week of month
```

```python
# Shifting ranges
ts = pd.Series(np.random.rand(10), index=pd.date_range('2018-01-01', periods=10))
ts
pd.concat([ts, ts.shift(1), ts.shift(2), ts.shift(3), ts.shift(4), ts.shift(5), ts.shift(6)], axis=1)
pd.concat([ts,ts.shift(-1),  ts / ts.shift(-1) - 1], axis=1)
pd.concat([ts, ts.shift(1, freq='D')], axis=1, ignore_index=True)
```

```python
# SHIFTING DATES USING OFFSET
from pandas.tseries.offsets import Day, MonthEnd
now = pd.datetime(2018, 1, 1)
now + 3 * Day()
now + MonthEnd()
now + MonthEnd(12)
```

```python
# RESAMPLING 
data.resample('Y')['Close'].mean()
data['2000': '2010'].resample('Y', kind='period').mean()
```

```python
# DOWNSAMPLING 
ts = pd.Series(np.random.rand(20), index=pd.date_range('2018-01-01', periods=20, freq='T'))
ts.resample('5min').mean()
ts.resample('5min', closed='right', label='right').mean()
```

```python
# OHLC RESAMPLING 
data.resample('Y', kind='period')['Close'].ohlc().head()
```

```python
# DownSAMPLING 
data_by_year = data.resample('Y', kind='period')['Close'].ohlc()
data_by_year.resample('Q', kind='period').asfreq()
data_by_year.resample('Q', kind='period').ffill(limit=1)
```

```python
plt.style.use('classic')
data['2006':].resample('B')['Close'].ffill().plot()
data['2006':].resample('B')['Close'].ffill().rolling(250).mean().plot()
data['2006':].resample('B')['Close'].ffill().rolling(250, min_periods=10).std().plot()
```
